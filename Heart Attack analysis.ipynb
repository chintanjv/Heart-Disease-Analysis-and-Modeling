## Data pre-processing

### Extract independent and target variables

### Independent variables

```python
x = df.drop(['heart_disease'], axis = 1)
x.head()
```

### Target variable

```python
y = df[['heart_disease']]
y.head()
```

### Splitting the data into train and test set

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
x_test
```

## Building and training the model

```python
from sklearn.metrics import accuracy_score
```

#### Decision tree classifier

```python
from sklearn.tree import DecisionTreeClassifier
r = DecisionTreeClassifier()
r.fit(x_train, y_train)
DecisionTree = r.predict(x_test)
DecisionTree
```

```python
accuracy_score(DecisionTree, y_test)
```

#### Random Forest classifier

```python
from sklearn.ensemble import RandomForestClassifier
k = RandomForestClassifier()
k.fit(x_train, y_train)
RandomForest = k.predict(x_test)
RandomForest
```

```python
accuracy_score(RandomForest, y_test)
```

#### Naive Bayes Classifier

```python
from sklearn.naive_bayes import GaussianNB
m = GaussianNB()
m.fit(x_train, y_train)
NaiveBayes = m.predict(x_test)
NaiveBayes
```

```python
accuracy_score(NaiveBayes, y_test)
```

#### K-neighbor classifier

```python
from sklearn.neighbors import KNeighborsClassifier
n = KNeighborsClassifier()
n.fit(x_train, y_train)
KNeighbors = n.predict(x_test)
KNeighbors
```

```python
accuracy_score(KNeighbors, y_test)
```

#### Support Vector Machine

```python
from sklearn.svm import SVC
s = SVC()
s.fit(x_train, y_train)
SupportVector = s.predict(x_test)
SupportVector
```

```python
accuracy_score(SupportVector, y_test)
```

## Models' accuracy comparison

```python
print("Decision tree classifier: ", accuracy_score(DecisionTree, y_test))
print("Random Forest Classifier: ", accuracy_score(RandomForest, y_test))
print("Naive Bayes classifier: ", accuracy_score(NaiveBayes, y_test))
print("K-neighbor classifier: ", accuracy_score(KNeighbors, y_test))
print("Support Vector Classifier: ", accuracy_score(SupportVector, y_test))
```

### Machine Learning models

##### Separating independent variables

```python
x1 = df.drop(["heart_disease"], axis = 1)
```

```python
x1.head(10)
```

##### Separating target variable

```python
y1 = df[["heart_disease"]]
```

```python
y1.head()
```

```python
#Splitting the dataset into train and test set
from sklearn.model_selection import train_test_split
x1_train, x1_test, y1_train, y1_test = train_test_split(x1, y1, test_size=0.30)
```

#### Logistic regression

```python
#Importing Logistic Regression
from sklearn.linear_model import LogisticRegression

#Training and then testing the model
lr = LogisticRegression()
lr.fit(x1_train, y1_train)
pred_lr = lr.predict(x1_test)
```

```python
#Printing out the accuracy
acc_lr = accuracy_score(y1_test, pred_lr)
acc_lr
```

```python
#Confusion matrix
from sklearn.metrics import confusion_matrix

cm_lr = confusion_matrix(y1_test, pred_lr)

#Plotting the matrix
sns.heatmap(cm_lr/np.sum(cm_lr), annot = True, fmt = '0.2%', cmap = "Reds");
```

Type-1 Error (False Positive): 10.99%,
Type-2 Error (False Negative): 9.89%,
The rest of the values are correctly predicted

#### Decision tree classifier

```python
#Importing Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier

#Traiing and testing the model
dtc = DecisionTreeClassifier()
dtc.fit(x1_train, y1_train)
pred_dtc = dtc.predict(x1_test)
```

```python
#Accuracy of the model
acc_dtc = accuracy_score(y1_test, pred_dtc)
acc_dtc
```

```python
#Confusion matrix
cm_dt = confusion_matrix(y1_test, pred_dtc)

#Plotting the matrix
sns.heatmap(cm_dt/np.sum(cm_dt), annot = True, fmt = '0.2%', cmap = "Reds")
```

Type-1 Error (False Positive): 12.09%,
Type-2 Error (False Negative): 10.99%,
The rest of the values are correctly predicted

#### Random Forests

```python
#Importing library
from sklearn.ensemble import RandomForestClassifier

#Creating model object
rf = RandomForestClassifier()

#Training model
rf.fit(x1_train, y1_train)

#Making Prediction
pred_rf = rf.predict(x1_test)
```

```python
#Accuracy of the model
acc_rf =  accuracy_score(y1_test, pred_rf)
acc_rf
```

```python
#Confusion matrix
cm_rf = confusion_matrix(y1_test, pred_rf)

#Plotting the matrix
sns.heatmap(cm_rf/np.sum(cm_rf), annot = True, fmt = "0.2%", cmap = "Reds")
```

Type-1 Error (False Positive): 10.99%,
Type-2 Error (False Negative): 9.89%,
The rest of the values are correctly predicted

#### K neighbors

```python
#Importing the XGB library
from sklearn.neighbors import KNeighborsClassifier

#Creating model object
kn = KNeighborsClassifier()

#Training model
kn.fit(x1_train, y1_train)

#Making prediction
pred_kn = kn.predict(x1_test)
```

```python
#Accuracy of the model
acc_kn = accuracy_score(y1_test, pred_kn)
acc_kn
```

```python
#Confusoin matrix
cm_kn = confusion_matrix(y1_test, pred_kn)

#Plotting the matrix
sns.heatmap(cm_kn/np.sum(cm_kn), annot = True, fmt = '0.2%', cmap = 'Reds')
```

#### SVM Classifier

```python
#Importing SVM Classifier
from sklearn.svm import SVC

#Creating model object
svc = SVC()

#Training model data
svc.fit(x1_train, y1_train)

#Making predictions
pred_svc = svc.predict(x1_test)
```

```python
#Accuracy of the model
acc_svc = accuracy_score(y1_test, pred_svc)
acc_svc
```

```python
from sklearn.metrics import confusion_matrix
pred_svc_train = svc.predict(x1_train)
acc_svc_train = accuracy_score(y1_train, pred_svc_train)
cm_train_svc = confusion_matrix(y1_train, pred_svc_train)
cm_test_svc = confusion_matrix(y1_test, pred_svc)
print("Train Data CM: ")
print(cm_train_svc, '\n')
print("Test Data CM: ")
print(cm_test_svc)

output = pd.DataFrame({"Test Set": ['Train', 'Test'],
                      "Accuracy": [acc_svc_train, acc_svc],
                      "TP":[cm_train_svc[1][1], cm_test_svc[1][1]],
                      "TN":[cm_train_svc[0][0], cm_test_svc[0][0]],
                      "FP":[cm_train_svc[0][1], cm_test_svc[0][1]],
                      "FN":[cm_train_svc[1][0], cm_test_svc[1][0]]})

display(output)
```

Type-1 Error (False Positive): 12.09%,
Type-2 Error (False Negative): 4.40%,
The rest of the values are correctly predicted
